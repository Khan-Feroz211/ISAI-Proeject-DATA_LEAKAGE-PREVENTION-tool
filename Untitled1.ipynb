```python
import os
import pandas as pd

data_folder = './data/input'
data_file = os.path.join(data_folder, 'mlops_dataset.csv')

# Load CSV
df = pd.read_csv(data_file)

# Preview dataset
print("Dataset preview:")
print(df.head())
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np

# Assume df is available from previous cells.

# 1. Prepare X and y from the original dataframe
X_raw = df.iloc[:, :-1]  # All columns except the last are features
y_raw = df.iloc[:, -1]   # The last column is the target ('notes')

# 2. Preprocess y: Encode target if categorical
# The LabelEncoder `le` was already created in cell Bg4SAk9egMTL and was applied to the 'notes' column.
# Re-initializing and fitting it here to ensure it's available and consistent.
le = LabelEncoder()
y_processed = le.fit_transform(y_raw)
num_classes = len(le.classes_)

# 3. Preprocess X: Handle categorical and numerical features
# Define columns to drop and features to encode
# 'experiment_id', 'run_date' are typically identifiers or date strings that need specific handling or can be dropped for a simple model.
cols_to_drop = ['experiment_id', 'run_date']
X_features = X_raw.drop(columns=cols_to_drop)

# Identify remaining categorical and numerical columns
categorical_features = X_features.select_dtypes(include=['object']).columns
numerical_features = X_features.select_dtypes(include=['number']).columns

# Create a column transformer for preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('num', 'passthrough', numerical_features)
    ])

# Apply preprocessing to X
X_processed_transformed = preprocessor.fit_transform(X_features)

# If the output of ColumnTransformer is a sparse matrix, convert it to a dense array
# And ensure all elements are float for tensorflow
if hasattr(X_processed_transformed, 'toarray'):
    X_processed_transformed = X_processed_transformed.toarray().astype(np.float32)
else:
    X_processed_transformed = X_processed_transformed.astype(np.float32)

# 4. Split the processed data
X_train, X_test, y_train, y_test = train_test_split(X_processed_transformed, y_processed, test_size=0.2, random_state=42)

print("Training samples after preprocessing:", X_train.shape[0])
print("Testing samples after preprocessing:", X_test.shape[0])
print("X_train shape after preprocessing:", X_train.shape)

# 5. Define a simple model
# Adjust input_shape based on the new X_train.shape[1]
# Adjust output layer and loss function based on the number of classes in y
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax') # Multi-class classification
])

# Loss function for multi-class classification
loss_function = 'sparse_categorical_crossentropy'

model.compile(optimizer='adam',
              loss=loss_function,
              metrics=['accuracy'])

# Train
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=32)
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.layers import LeakyReLU, BatchNormalization
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
import numpy as np
import pandas as pd
import os
import kagglehub
import shutil
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# --- Re-establish the environment (as previous state might have been lost) --- start
project_path = '/content/ISAI-Proeject-DATA_LEAKAGE-PREVENTION-tool'

# Ensure the current working directory is set
if os.getcwd() != project_path:
    # Only clone/chdir if not already in the correct directory
    if not os.path.exists(project_path):
        !git clone https://github.com/Khan-Feroz211/ISAI-Proeject-DATA_LEAKAGE-PREVENTION-tool.git
    os.chdir(project_path)
print("Working directory:", os.getcwd())

# Make sure ai_models exists and has __init__.py
ai_models_path = os.path.join(project_path, 'ai_models')
if not os.path.exists(ai_models_path):
    os.makedirs(ai_models_path)
init_file = os.path.join(ai_models_path, '__init__.py')
if not os.path.exists(init_file):
    open(init_file, 'w').close()  # create empty __init__.py

# Download dataset (use Colab cache if available for faster access)
dataset_path = kagglehub.dataset_download("daniilkrizhanovskyi/mlops-dataset")

# Create project data directory and copy dataset
project_data_path = './data/input'
os.makedirs(project_data_path, exist_ok=True)

# Copy CSV file to project data path
csv_files = [f for f in os.listdir(dataset_path) if f.endswith('.csv')]
if csv_files:
    data_file = os.path.join(dataset_path, csv_files[0])
    shutil.copy(data_file, project_data_path)
    data_file = os.path.join(project_data_path, csv_files[0])
    df = pd.read_csv(data_file)
    print("Dataset loaded successfully.")
else:
    raise FileNotFoundError("No CSV file found in the downloaded dataset.")
# --- Re-establish the environment --- end

# 1. Prepare X and y from the original dataframe
X_raw = df.iloc[:, :-1]  # All columns except the last are features
y_raw = df.iloc[:, -1]   # The last column is the target ('notes')

# 2. Preprocess y: Encode target if categorical
le = LabelEncoder()
y_processed = le.fit_transform(y_raw)
num_classes = len(le.classes_)

# 3. Preprocess X: Handle categorical and numerical features
cols_to_drop = ['experiment_id', 'run_date']
X_features = X_raw.drop(columns=cols_to_drop)

categorical_features = X_features.select_dtypes(include=['object']).columns
numerical_features = X_features.select_dtypes(include=['number']).columns

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('num', StandardScaler(), numerical_features)
    ])

X_processed_transformed = preprocessor.fit_transform(X_features)

if hasattr(X_processed_transformed, 'toarray'):
    X_processed_transformed = X_processed_transformed.toarray().astype(np.float32)
else:
    X_processed_transformed = X_processed_transformed.astype(np.float32)

# 4. Split the processed data with stratification
X_train, X_test, y_train, y_test = train_test_split(X_processed_transformed, y_processed, test_size=0.2, random_state=42, stratify=y_processed)
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")

# 5. Calculate class weights
class_weights_array = class_weight.compute_class_weight(
    'balanced', classes=np.unique(y_processed), y=y_processed
)
class_weights = {i : class_weights_array[i] for i in range(len(class_weights_array))}

print("Data preprocessing and class weights calculated successfully with stratified split.")

# 6. Define model_v6 (same as before)
model_v6 = tf.keras.Sequential([
    tf.keras.layers.Dense(128, input_shape=(X_train.shape[1],)),
    BatchNormalization(),
    LeakyReLU(alpha=0.01),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(64),
    BatchNormalization(),
    LeakyReLU(alpha=0.01),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(32),
    BatchNormalization(),
    LeakyReLU(alpha=0.01),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(16),
    BatchNormalization(),
    LeakyReLU(alpha=0.01),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(num_classes, activation='softmax')
])

# 7. Compile model_v6
model_v6.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("\nModel V6 redefined and compiled successfully with LeakyReLU, BatchNormalization, and increased layers.")
model_v6.summary()

# 8. Train model_v6 with class_weights
history_v6 = model_v6.fit(X_train, y_train,
                         validation_data=(X_test, y_test),
                         epochs=100,
                         batch_size=32,
                         class_weight=class_weights,
                         verbose=0)

# 9. Evaluate the trained model's performance
loss_v6, accuracy_v6 = model_v6.evaluate(X_test, y_test, verbose=0)
print(f'\nModel V6 Test Loss: {loss_v6:.4f}')
print(f'Model V6 Test Accuracy: {accuracy_v6:.4f}')

# Get predictions from model_v6 on the test set
y_pred_probabilities_v6 = model_v6.predict(X_test)
y_pred_v6 = np.argmax(y_pred_probabilities_v6, axis=1)

print("\n--- Model V6 Classification Report after Stratified Split ---")
print(classification_report(y_test, y_pred_v6, target_names=le.classes_, zero_division=0))

# Generate Confusion Matrix
cm_v6 = confusion_matrix(y_test, y_pred_v6)
plt.figure(figsize=(10, 8))
sns.heatmap(cm_v6, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Model V6 Confusion Matrix after Stratified Split')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# 10. Plot training & validation accuracy values
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history_v6.history['accuracy'])
plt.plot(history_v6.history['val_accuracy'])
plt.title('Model V6 Accuracy (Stratified Split)')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# 11. Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history_v6.history['loss'])
plt.plot(history_v6.history['val_loss'])
plt.title('Model V6 Loss (Stratified Split)')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.tight_layout()
plt.show()
```
